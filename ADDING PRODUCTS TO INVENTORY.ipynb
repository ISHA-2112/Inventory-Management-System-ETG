```python
import json
fh=open('record.json')
records=json.load(fh) #reading the record from file
while True:
    try:
        update=input("\nDo you want to update an existing product? (please enter 'yes' or 'no'): ")
        update=update.lower() #even if user enters YES, Yes or yes it is acceptable
        if update=='yes':
            prod_id=input('Enter the product id: ')
            r=int(prod_id) #to check if product id given is an integer
            quan=input('Enter the quantity: ')
            quan=int(quan)
            records[prod_id]['quan']=records[prod_id]['quan']+quan #increases the quantity in the inventory
            loop=input("\nDo you want to enter more products?(enter'yes' or 'no'): ")
            loop=loop.lower() #even if user enters YES, Yes or yes it is acceptable 
            if loop=='yes':
                continue
            elif loop=='no':
                break #program complete
            
        elif update=='no':
            prod_id=str(len(records)+1) #enters the next product id on its own
            name=input('Enter the name of product: ')
            price=int(input('Enter the price of product: '))
            quan=input('Enter the quantity: ')
            quan=int(quan)
            exp_date=input('Enter the expiry date: ')
            net_wt=input('Enter the net weight: ')
            records[prod_id]={'name':name,'price':price,'quan':quan,'exp_date':exp_date,'net_wt':net_wt}
                     #adds new product to inventory
            
            lp=input("\nDo you want to enter more products?(enter'yes' or 'no') ")
            lp=lp.lower() #even if user enters YES, Yes or yes it is acceptable
            if lp=='yes':
                continue
            elif lp=='no':
                break #program complete
            
        else:
            print('Please enter yes/no') #if some wrong input is given
    except:
        print('Oops! please try again') #if integer input is not for product id or for quantity 
fs=json.dumps(records)
ff=open('record.json','w') #writing the new record back in file
ff.write(fs)
ff.close()
        
```
